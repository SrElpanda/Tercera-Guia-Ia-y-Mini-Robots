Importar las librerías necesarias

```python
from gtts import gTTS
from IPython.display import Audio
from wordfreq import random_words
import difflib
import random
```

Inicialización de variables

```python
palabra_objetivo = "perro"
n_poblacion = 50
prob_mutacion = 0.5 #Alta probabilidad de mutación para reducir la probabilidad de que el algorítmo se estanque
generaciones = 120
```

Función que generar el archivo mp3 y el reproductor para la reproducción por voz de las palabras

```python
def voz(poblacion):
  palabras = " ".join(poblacion)
  tts = gTTS(text=palabras, lang="es")
  tts.save("voz.mp3")
  display(Audio("voz.mp3"))
```

Generación de la población de 50 palabras aleatorias

```python
def pob_init():
  palabras = random_words(lang='es', nwords=n_poblacion)
  return palabras.split()
```

Función que calcula la aptitud según una palabra objetivo, para la comparación se hace uso del método sequenceMatcher de difflib

```python
def aptitud(palabra,objetivo):
  return difflib.SequenceMatcher(None,palabra,objetivo).ratio() #el ratio va de 0 a 1
```

Función que realiza la selección de 2 padres por probabilidad según aptitud

```python
def seleccion(palabras):
  aptitudes = []
  for palabra in palabras:
    aptitudes.append(aptitud(palabra,palabra_objetivo))
  return random.choices(palabras,aptitudes,k=2)
```

Función que realiza el cruce entre los 2 padres seleccionados

```python
def cruce(padre1,padre2):
  k = min(len(padre1), len(padre2)) // 2
  hijo = padre1[:k] + padre2[k:]
  return hijo
```

Función que realiza la mutación según la probabilidad de mutación, la mutación se da al cambiar la palabra por alguna otra aleatoria

```python
def mutacion(prob_mutacion,palabra):
  if random.random() <= prob_mutacion:
    return random_words(lang='es', nwords=1)
  return palabra
```

ejecución principal del algoritmo Genético

```python
print("Población Inicial")
poblacion = pob_init()
voz(poblacion)
print("\n")
print("Palabra objetivo: ",palabra_objetivo)
for i in range(generaciones):

  poblacion = sorted(poblacion, key=lambda x: aptitud(x, palabra_objetivo), reverse=True)
  mejor = poblacion[0]

  print("Generación ",i,"el mejor cromosoma es ",mejor," con ",aptitud(mejor,palabra_objetivo)*100,"% de similitud")


  #En caso de querer la reproducción por voz de toda la población en todas las generaciones, descomentar las siguientes linea (se demora más la busqueda)
  #print("Reproducción por voz de la población: ")
  #voz(poblacion)



  if aptitud(mejor,palabra_objetivo) == 1:
    print("Se ha encontrado la palabra objetivo en la generación ",i)
    break

  nueva_poblacion = poblacion[:5] # se Realiza selección de algunos cromosomas por elitismo

  while len(nueva_poblacion) < n_poblacion:
    padres = seleccion(poblacion)
    hijo = cruce(padres[0],padres[1])
    hijo = mutacion(prob_mutacion,hijo)
    nueva_poblacion.append(hijo)

  poblacion = nueva_poblacion
print("Población Final")
voz(poblacion)
```
